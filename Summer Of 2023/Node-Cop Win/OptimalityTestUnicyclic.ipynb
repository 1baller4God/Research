```python
def is_corner(G,v):
    V = G.vertices();
    N = set(G.neighbors(v,closed=True));
    for u in V:
        if u != v:
            M = set(G.neighbors(u,closed=True));
            if N.issubset(M):
                return True
    return False
```

```python
def is_copwin(G):
    H = Graph(G);
    V = H.vertices();
    n = len(V);
    if n == 1:
        return True
    for u in V:
        if is_corner(H,u):
            H.delete_vertex(u);
            return is_copwin(H)
    return False
```

```python
def NCRel(G,p):
    P=0;
    V=G.vertices();
    n = len(V);
    C=Combinations(V).list()
    for U in C:
        H=G.subgraph(U)
        if is_copwin(H):
            k=H.order()
            P+=p^k*(1-p)^(n-k)
    return P
```

```python
def CS(G,x):
    P=0;
    V=G.vertices();
    n = len(V);
    C=Combinations(V).list()
    for U in C:
        H=G.subgraph(U)
        if H.is_connected():
            k=H.order()
            P+=x^k
    return P
```
