```python
from sage.all import *




# Create the circulant graph
def circulant_graph(number_of_vertices):
    
    connections = []
    
    for vertex in range(number_of_vertices):

        for new_vertex in range(number_of_vertices - 1):
            
            if vertex != new_vertex:
            
                connection = [vertex, new_vertex]
                
                connections.append(connection)
        
    return connections

cc = circulant_graph(12)
print(cc)

show(Graph(cc))
```

```python
def circulant(n, L):
    E = []
    
    for i in range(n):
        for j in range(i+1, n):
            if(((i-j) % n) in  L):
                if({i,j} not in E):
                    E.append({i,j})
                    
    G = Graph(E)
    return(G)
```

```python
G = circulant(9,[3,4,5,6])
```

```python
show(G, layout='circular')
```

```python
G.show(layout='circular')
```
